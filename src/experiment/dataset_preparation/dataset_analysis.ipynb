```python
# remove 50% of df
df = df.iloc[:len(df)//2]
```

```python
y = df["sentiment"]
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, :1]
```

```python
X = df["text"]
```

```python
# remove title column from X
X = X.drop(columns=["Title"])
X = X.drop(columns=["Description"])
```

```python
# remove samples with missing values from X,y
X = X.dropna()
y = y.loc[X.index]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Step 1: Reduce vocabulary size roughly
# before: max=10000
# 20, 0.8
# nyt-comments-april17, all-in-one_sentiment: 50, 0.8
vectorizer = CountVectorizer(binary=True, min_df=50, max_df=0.8)
X = vectorizer.fit_transform(X.values)
```

```python
# Step 2: Select best
from sklearn.feature_selection import SelectKBest, chi2
X = SelectKBest(chi2, k=1006).fit_transform(X, y)
```

```python
# transform X to df
X = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out()[:1006])
```

# Reduce class indexes by 1

```python
#check if any label "0" is present in y
if (y == 0).any().any():
    print("Label '0' is present in y")
else:
    print("Label '0' is not present in y")
```

```python
# reduce all class indexes in y by 1
y = y - 1
```

# Remove gaps in labels

```python
# remove gaps in labels, as in if there is a label "0" and then a label "2", turn the 2 into a 1
y = y.apply(lambda x: pd.factorize(x)[0])
```

# Dataset info

```python
# number of samples
sample_count = len(X)
sample_count
```

```python
# number of classes
class_count = len(np.unique(y.values))
class_count
```

```python
# number of features
feature_count = X.shape[1]
feature_count
```

```python
# Convert y to int if needed
class_counts = y.value_counts()
print(class_counts)
```

```python
class_percentages_sorted = get_sorted_class_percentages_label_encoded(y)
print(class_percentages_sorted)
```

```python
highest_class_percentage = class_percentages_sorted.iloc[0]
lowest_class_percentage = class_percentages_sorted.iloc[-1]
```

```python
class_percentage_range = highest_class_percentage - lowest_class_percentage
print(f"Difference between highest and lowest class percentage: {round(class_percentage_range, 2)} % point")
```

```python
if lowest_class_percentage != 0:
    ratio = highest_class_percentage / lowest_class_percentage
    print(f"The highest class percentage is {round(ratio, 2)} times higher than the lowest.")
else:
    ratio = np.nan
    print("The lowest class percentage is 0, cannot calculate ratio.")
```

```python
average_class_percentage = class_percentages_sorted.mean()
print(f"Average class count percentage: {round(average_class_percentage, 3)} %")
```

```python
class_percentages_std = class_percentages_sorted.std()
print(f"Standard deviation of class percentages: {round(class_percentages_std, 3)} %")
```

```python
cov = class_percentages_std / average_class_percentage
print(f"Coefficient of variation of class percentages: {round(cov, 3)}")
```

# Save class percentages, dataset info and X, y CSVs

```python
# Include classes with 0 in class_percentages.csv (for analysis purposes)
class_percentages_sorted.to_csv(DATASET_DIR / "class_percentages.csv", header=False)
```

```python
summary_stats = {
    "sample_count_total": sample_count,
    "class_count": class_count,
    "highest_class_percentage": highest_class_percentage,
    "lowest_class_percentage": lowest_class_percentage,
    "class_percentage_range": class_percentage_range,
    "class_percentage_times_higher": ratio,
    "average_class_percentage": average_class_percentage,
    "class_percentages_std": class_percentages_std,
    "cov": cov
}

summary_df = pd.DataFrame([summary_stats])
summary_df.to_csv(DATASET_DIR / "class_stats_summary.csv", index=False)
```
