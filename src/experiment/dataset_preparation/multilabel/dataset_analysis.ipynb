# Dataset info

```python
# number of samples
sample_count = len(X)
sample_count
```

```python
# number of classes
class_count = y.shape[1]
class_count
```

```python
# Convert y to int if needed
y_int = y.astype(int)
class_counts = y_int.sum(axis=0).sort_values(ascending=False)
print(class_counts)
```

```python
class_percentages_sorted = get_sorted_class_percentages(y)
print(class_percentages_sorted)
```

```python
# Remove classes with 0.000 percentage
nonzero_classes = class_percentages_sorted[class_percentages_sorted > 0].index
y = y[nonzero_classes]
class_count_nonzero = y.shape[1]
print(f"Number of classes after removal of labels with no true samples: {class_count_nonzero}")

# Update class_percentages after removing zero-percentage classes
class_percentages_sorted_nonzero = class_percentages_sorted[nonzero_classes]
```

```python
# Check how many samples don't have any labels
no_labels = y.astype(int).sum(axis=1) == 0
no_labels_count = no_labels.sum()
only_samples_with_labels_count = sample_count - no_labels_count
print(f"Number of samples without any labels: {no_labels_count}")
print(f"Number of samples left: {only_samples_with_labels_count}")

# show samples with no labels
if no_labels_count > 0:
    no_labels_array = y[no_labels]

# Reset index for X and y after filtering out samples without labels
X = X.reset_index(drop=True)
y = y.reset_index(drop=True)

# Remove samples without any labels
X_only_samples_with_labels = X[~no_labels].reset_index(drop=True)
y_only_samples_with_labels = y[~no_labels].reset_index(drop=True)
```

```python
highest_class_percentage = class_percentages_sorted_nonzero.iloc[0]
lowest_class_percentage = class_percentages_sorted_nonzero.iloc[-1]
```

```python
class_percentage_range = highest_class_percentage - lowest_class_percentage
print(f"Difference between highest and lowest class percentage: {round(class_percentage_range, 2)} %")
```

```python
highest = class_percentages_sorted_nonzero.iloc[0]
lowest = class_percentages_sorted_nonzero.iloc[-1]
if lowest != 0:
    ratio = highest / lowest
    print(f"The highest class percentage is {round(ratio, 2)} times higher than the lowest.")
else:
    ratio = np.nan
    print("The lowest class percentage is 0, cannot calculate ratio.")
```

```python
average_class_percentage = class_percentages_sorted_nonzero.mean()
print(f"Average class count percentage: {round(average_class_percentage, 3)} %")
```

```python
class_percentages_std = class_percentages_sorted_nonzero.std()
print(f"Standard deviation of class percentages: {round(class_percentages_std, 3)} %")
```

```python
# Cardinality: average number of labels per sample
cardinality = get_cardinality(y)
print("Cardinality:", round(cardinality, 3))
```

```python
# Cardinality: average number of labels per sample
cardinality_only_samples_with_labels = get_cardinality(y_only_samples_with_labels)
print("Cardinality:", round(cardinality_only_samples_with_labels, 3))
```

# Save class percentages, dataset info and X, y CSVs

```python
# Include classes with 0 in class_percentages.csv (for analysis purposes)
class_percentages_sorted.to_csv(DATASET_DIR / "class_percentages.csv", header=False)
```

```python
summary_stats = {
    "sample_count_total": sample_count,
    "only_samples_with_labels_count": only_samples_with_labels_count,
    "class_count": class_count,
    "highest_class_percentage": highest_class_percentage,
    "lowest_class_percentage": lowest_class_percentage,
    "class_percentage_range": class_percentage_range,
    "class_percentage_times_higher": ratio,
    "average_class_percentage": average_class_percentage,
    "class_percentages_std": class_percentages_std,
    "cardinality": cardinality,
    "cardinality_only_samples_with_labels": cardinality_only_samples_with_labels
}

summary_df = pd.DataFrame([summary_stats])
summary_df.to_csv(DATASET_DIR / "class_stats_summary.csv", index=False)
```
