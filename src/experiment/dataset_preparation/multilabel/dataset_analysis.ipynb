# Dataset loading (arff)

```python
FILENAME =  NAME + ".arff"

with open(DATASET_DIR / FILENAME, 'r') as f:
    dataset = arff.load(f)
```

```python
df = pd.DataFrame(dataset['data'], columns=[attr[0] for attr in dataset['attributes']])
df
```

```python
X = df.iloc[:, :-LABEL_COUNT]  # Features
y = df.iloc[:, -LABEL_COUNT:]  # Labels
```

# Dataset info

```python
# number of samples
sample_count = len(X)
sample_count
```

```python
# number of classes
class_count = y.shape[1]
class_count
```

```python
class_percentages_sorted = get_sorted_class_percentages(y)
print(class_percentages_sorted)
```

```python
# Remove classes with 0.000 percentage
nonzero_classes = class_percentages_sorted[class_percentages_sorted > 0].index
y_nonzero = y[nonzero_classes]
print(f"Number of classes after removal of labels with no true samples: {y_nonzero.shape[1]}")

# Update class_percentages after removing zero-percentage classes
class_percentages_sorted_nonzero = class_percentages_sorted[nonzero_classes]
```

```python
class_percentage_range = class_percentages_sorted_nonzero.iloc[0] - class_percentages_sorted_nonzero.iloc[-1]
print(f"Difference between highest and lowest class percentage: {round(class_percentage_range, 2)} %")
```

```python
highest = class_percentages_sorted_nonzero.iloc[0]
lowest = class_percentages_sorted_nonzero.iloc[-1]
if lowest != 0:
    ratio = highest / lowest
    print(f"The highest class percentage is {round(ratio, 2)} times higher than the lowest.")
else:
    ratio = np.nan
    print("The lowest class percentage is 0, cannot calculate ratio.")
```

```python
average_class_percentage = class_percentages_sorted_nonzero.mean()
print(f"Average class count percentage: {round(average_class_percentage, 3)} %")
```

```python
class_percentages_std = class_percentages_sorted_nonzero.std()
print(f"Standard deviation of class percentages: {round(class_percentages_std, 3)} %")
```

```python
# Cardinality: average number of labels per sample
cardinality = get_cardinality(y)
print("Cardinality:", round(cardinality, 3))
```

# Save class percentages, dataset info and X, y CSVs

```python
# Include classes with 0 in class_percentages.csv (for analysis purposes)
class_percentages_sorted.to_csv(DATASET_DIR / "class_percentages.csv", header=False)
```

```python
summary_stats = {
    "sample_count": sample_count,
    "class_count": class_count,
    "highest_class_percentage": class_percentages_sorted.iloc[0],
    "lowest_class_percentage": class_percentages_sorted.iloc[-1],
    "class_percentage_range": class_percentage_range,
    "class_percentage_times_higher": ratio,
    "average_class_percentage": average_class_percentage,
    "class_percentages_std": class_percentages_std,
    "cardinality": cardinality
}

summary_df = pd.DataFrame([summary_stats])
summary_df.to_csv(DATASET_DIR / "class_stats_summary.csv", index=False)
```
