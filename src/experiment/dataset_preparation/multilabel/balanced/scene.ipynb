# Train and Test (80%, 20%)

```python
from skmultilearn.model_selection import iterative_train_test_split

# Convert to NumPy arrays if they are DataFrames
X_np = X_copy.values
y_np = y_copy.values

X_train, y_train, X_test, y_test = iterative_train_test_split(X_np, y_np, test_size=0.2)

X_train_df = pd.DataFrame(X_train, columns=X_copy.columns)
y_train_df = pd.DataFrame(y_train, columns=y_copy.columns)
X_test_df = pd.DataFrame(X_test, columns=X_copy.columns)
y_test_df = pd.DataFrame(y_test, columns=y_copy.columns)
```
