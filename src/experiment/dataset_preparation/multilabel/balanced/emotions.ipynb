```python
NAME = "emotions"
```

```python
FILENAME =  NAME + ".arff"
DATASET_DIR = dataset_root_dir / 'multilabel' / NAME
print(FILENAME)
print(DATASET_DIR)
```

```python
trimmed_name = NAME + "_trimmed"
DATASET_DIR_TRIMMED = dataset_root_dir / 'multilabel' / trimmed_name
print(DATASET_DIR_TRIMMED)
```

# Dataset loading

```python
with open(DATASET_DIR / FILENAME, 'r') as f:
    dataset = arff.load(f)
```

```python
df = pd.DataFrame(dataset['data'], columns=[attr[0] for attr in dataset['attributes']])
df
```

```python
X = df.iloc[:, :-6]  # Features
y = df.iloc[:, -6:]  # Labels
```

```python
X
```

```python
y
```

# Dataset info

```python
# number of samples
len(X)
```

```python
# Convert y to numeric
y_numeric = y.apply(pd.to_numeric)

# Calculate percentage for each class
class_percentages = y_numeric.sum() / len(y_numeric) * 100
print(class_percentages)
```

```python
# Cardinality: average number of labels per sample
cardinality = y.apply(pd.to_numeric).sum(axis=1).mean()
print("Cardinality:", cardinality)
```

# Trimming

Repeated many times

```python
protected_classes = ['happy-pleased', 'quiet-still', 'sad-lonely']
X, y = trim_class(X, y, "angry-aggresive", 1, protected_classes)
print("Total count of samples after trimming:", len(X))
```

```python
protected_classes = ['happy-pleased', 'quiet-still']
X, y = trim_class(X, y, "relaxing-calm", 1, protected_classes)
print("Total count of samples after trimming:", len(X))
```

```python
X.to_csv(DATASET_DIR_TRIMMED / "X.csv", index=False)
y.to_csv(DATASET_DIR_TRIMMED / "y.csv", index=False)
```
