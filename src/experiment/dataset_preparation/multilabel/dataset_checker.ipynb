```python
from pathlib import Path
import arff
import pandas as pd
import numpy as np
from src.experiment.helpers.variables import dataset_root_dir
```

```python
LABEL_COUNT = 159
NAME = "bibtex"
```

```python
FILENAME =  NAME + ".arff"
DATASET_DIR = dataset_root_dir / 'multilabel' / NAME
print(DATASET_DIR)
```

# Dataset loading

```python
with open(DATASET_DIR / FILENAME, 'r') as f:
    dataset = arff.load(f)
```

```python
df = pd.DataFrame(dataset['data'], columns=[attr[0] for attr in dataset['attributes']])
df
```

```python
X = df.iloc[:, :-LABEL_COUNT]  # Features
y = df.iloc[:, -LABEL_COUNT:]  # Labels
```

```python
X
```

```python
y
```

# Dataset info

```python
# number of samples
len(X)
```

```python
# Convert y to numeric
y_numeric = y.apply(pd.to_numeric)

# Calculate percentage for each class
class_percentages = y_numeric.sum() / len(y_numeric) * 100
print(class_percentages)
```

```python
class_percentages_sorted = class_percentages.sort_values(ascending=False)
print(class_percentages_sorted)
```

```python
class_percentages_sorted.to_csv(DATASET_DIR / "class_percentages.csv", header=False)
```

```python
# Cardinality: average number of labels per sample
cardinality = y.apply(pd.to_numeric).sum(axis=1).mean()
print("Cardinality:", round(cardinality, 3))
```

```python
average_class_percentage = class_percentages.mean()
print(f"Average class count percentage: {round(average_class_percentage, 3)} %")
```

```python
class_percentages_std = class_percentages.std()
print(f"Standard deviation of class percentages: {round(class_percentages_std, 3)} %")
```
